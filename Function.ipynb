```python
def cal(x,y):
    a = x*y
    return a
    
l = int(input("Enter l = "))  #global variable
b = int(input("Enter b = "))
h = int(input("Enter h = "))   
    
# print(cal(l,b,h))
ar = cal(l,b)

volume = ar*h

print(volume)
```

```python
def area(x,y):
    a = x*y
    return a

def volume(x,y,z):
    v = area(x,y)*z
    return v
    
l = int(input("Enter l = "))  #global variable
b = int(input("Enter b = "))
h = int(input("Enter h = "))   
    
# print(cal(l,b,h))
vol = volume(l,b,h)
print(vol)
```
